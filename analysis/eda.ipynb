# An√°lisis de los datasets

```python
import pandas as pd
```

```python
files_path = [
    '../downloads/disney_plus_titles.csv',
    '../downloads/netflix_titles.csv'
]
```

```python
dfs = []

for index, file_path in enumerate(files_path):
    if index == 0:
        df = pd.read_csv(file_path, sep=',')
        platform = 'DP'
    else:
        df = pd.read_csv(file_path, sep=';')
        platform = 'N'

    df['platform'] = platform

    dfs.append(df)
```

```python
dfs_concat = pd.concat([df for df in dfs], ignore_index=True)
```

```python
dfs_concat.info()
```

```python
dfs_concat.isnull().sum()
```

Como se puede observar el dataset presenta valores faltantes en determinados campos. La mayor cantidad se encuentra en los campos de "director", "cast" y "country".
